```python
from bokeh.io import output_notebook
from bokeh.plotting import figure, output_file, output_notebook, show
import pandas as pd
import numpy as np
output_notebook()


from bokeh.models import ColumnDataSource, HoverTool
from bokeh.sampledata.autompg import autompg_clean as df
from bokeh.transform import factor_cmap

# Abrimos el archivo csv y lo agregamos en un DataFrame
dataFrame01 = pd.read_csv("vehiculos.csv", sep=",", engine='python')

# Declaramos una paleta de colores
colores = ['#2C638B', '#2C8B52', '#8B3D2C', '#2C8B6E','#000000']

group = dataFrame01.groupby('MARCA', as_index = False).mean()

x_range= group['MARCA']
y_range= group['MODELO']

figure01 = figure(x_range=x_range, plot_height=550, title="MODELO MARCA DE LOS VEHICULOS", toolbar_location=None, tools="")
figure01.vbar(x=x_range, top=group['MODELO'], width=0.9, line_color=colores[4], line_width=2, fill_color=colores[:2])
figure01.y_range.start = 1999
figure01.yaxis.axis_label = "MODELO"
figure01.xaxis.axis_label = "MARCA"
figure01.xaxis.major_label_orientation = 1.2
show(figure01)
```

```python
from bokeh.io import output_notebook
from bokeh.plotting import figure, output_file, output_notebook, show
import pandas as pd
import numpy as np
output_notebook()


from bokeh.models import ColumnDataSource, HoverTool
from bokeh.sampledata.autompg import autompg_clean as df
from bokeh.transform import factor_cmap

# Abrimos el archivo csv y lo agregamos en un DataFrame
dataFrame01 = pd.read_csv("vehiculos.csv", sep=",", engine='python')

# Declaramos una paleta de colores
colores = ['#2C638B', '#2C8B52', '#8B3D2C', '#2C8B6E','#000000']

group = dataFrame01.groupby('CLASE', as_index = False).mean()

x_range= group['CLASE']
y_range= group['ASIENTOS']

figure01 = figure(x_range=x_range, plot_height=550, title="ASIENTOS POR CLASE DE VEHICULO", toolbar_location=None, tools="")
figure01.vbar(x=x_range, top=group['ASIENTOS'], width=0.9, line_color=colores[4], line_width=2, fill_color=colores[:3])
figure01.y_range.start = 0
figure01.yaxis.axis_label = "ASIENTOS"
figure01.xaxis.axis_label = "CLASE"
figure01.xaxis.major_label_orientation = 1.2
show(figure01)
```

```python

import pandas as pd
from bokeh.plotting import figure, output_file, show
from bokeh.models import ColumnDataSource
from bokeh.models.tools import HoverTool


#Abrimos el archivo y loa agregamos a un dataframe
df =pd.read_csv("vehiculos.csv", sep=",", engine='python')

sample = df.sample(4000)
source = ColumnDataSource(sample)

#Se determina que variale sera las x y cual las y
p = figure()
p.circle(x='ASIENTOS', y='MODELO', 
         source=source, 
         size=10, color='blue')

p.title.text = 'MODELO DEL VEHICULO'
p.xaxis.axis_label = 'ASIENTOS'
p.yaxis.axis_label = 'MODELO'


hover = HoverTool()
hover.tooltips=[
    ('CLASE', '@CLASE')
]

p.add_tools(hover)

show(p)
```
